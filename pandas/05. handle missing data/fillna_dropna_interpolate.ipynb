```python
# To set day as an index

df = pd.read_csv("weather_data.csv", parse_dates=["day"])
df.set_index('day', inplace=True)
df 
```

<span style="color:blue">
    <h3>fillna</h3> 
</span>
<span style="color:violet">
    Fill all NaN with one specific value
</span>

```python
# Use fillna() method in dataframe. 
# Special handling. 
# Missing data replaced NaN with 0 value. 
new_df = df.fillna(0)
new_df
```

<span style="color:violet">
    Fill na using column names and dict
</span>

```python
# Different values for different columns
# Inside dictionary pass new values for particular columns

new_df = df.fillna({
    'temperature': 0,
    'windspeed': 0,
    'event': 'no event',    
})
new_df
```

<span style="color:violet">
    Use method to determine how to fill na values
</span>

```python
# forward fill ffill
# carry forward the value to next cell
# previous method was
# df.fillna(method="ffill")
# now its df.ffill() forward fill
# df.bfill() Backward fill

new_df = df.ffill()
new_df
```

<span style="color:violet">
    Use of axis
</span>

```python
# Using axis copying data horizontally on other cells

import pandas as pd
new_df = df.bfill(axis="columns")
new_df
```

<span style="color:violet">Limit parameter</span>

```python
# copying cell output in certain cells using limit parameter

new_df = df.ffill(limit=1)
new_df
```

<span style="color:violet">Interpolate</span>

```python
"""
infer_objects(): Tries to convert object columns to their best-matching types (like int, float, etc.).

interpolate(): Fills in missing values using interpolation (linear, spline, etc.).
"""

new_df = df.infer_objects()
new_df
```

```python

new_df = df.infer_objects(copy=False)
new_df = df.interpolate(method="time")
new_df
```

<span style="color:violet">Dropna</span>

```python
# dropna() in pandas â€” Think of it as a cleaning tool 
# to remove missing data (NaN values) 
# from your DataFrame.
# prints out the valid content and remove NaN values

new_df = df.dropna()
new_df
```

```python
# It will drop rows where all values are NaN. 
# In other words, it removes rows that are completely empty 
# (i.e., all columns in that row have missing values).
new_df = df.dropna(how="all")
new_df
```

```python
# keep the row if I have only Non Na value

new_df = df.dropna(thresh=1)
new_df
```
