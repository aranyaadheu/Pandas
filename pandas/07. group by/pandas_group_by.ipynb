The topic is to look at weather data from various cities and see how group by can be used to run some analytics.


```python
import pandas as pd 
df = pd.read_csv("weather_by_cities.csv")
df
```

<h2>For this dataset, lets try to look up at the following answers,</h2>

1. What was the maximum temp in each of these 3 cities?

2. What was the average windspeed in each of these 3 cities?

```python
# dataframe groupby object
g = df.groupby("city")
g
```

**DataFrameGroupBy object looks something like below**


```python
from IPython.display import Image, display, HTML

HTML('''
<style>
div.input {display:none;}
</style>
''')

display(Image(filename='group_by_cities.png'))
```

```python
for city, city_df in g:
    print(city)
    print("\n")
    print(city_df)
```

**This is similar to SQL**

**SELECT** * **from weather_data GROUP BY city**

```python
# to access a specific dataframe
g.get_group("mumbai")
```

```python
# max temp of each cities
g.max()
```

```python
# average temp of each cities
# add numeric_value to True to get the avg value of the df
 
g.mean(numeric_only=True)
```
