<span style="color: blue;">
    <h2>Pivot Table</h2>
</span>

Pivot allows you to transform or reshape dataframe


```python
import pandas as pd  
import numpy as np 

df = pd.read_csv("weather.csv")
df 
```

```python
df.pivot(index='city' ,columns='date')
```

```python
# index as rows
df.pivot(index='date', columns='humidity')
```

```python
df.pivot(index='city', columns='date', values='humidity')
```

```python
df.pivot(index='date', columns='city')
```

```python
df.pivot(index='humidity', columns='city')
```

<span style="color: blue;">
    <h3>Pivot Table</h3>
</span>

Pivot table is used to summerize and aggregate data inside dataframe

```python
df = pd.read_csv("weather2.csv")
df
```

```python
df.pivot_table(index='city', columns='date')
```

<span style="color: red;">
    <h3>Margins</h3>
</span>

```python
# aggfunc sum is defined aggfunc="sum"

df.pivot_table(index="city", columns='date', margins=True, aggfunc="sum")
```

```python
df.pivot_table(index='city', columns='date', margins=True, aggfunc="count")
```

More usage of pandas pivot table in pandas documentaion (https://pandas.pydata.org/docs/reference/api/pandas.pivot_table.html)

<span style="color: red;">
    <h3>Grouper</h3>
</span>

```python
df = pd.read_csv("weather.csv")
df 
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
# monthly average temperature using grouper
# freq="ME"
# Month-End frequency
# apply date time index first

df.pivot_table(index=pd.Grouper(freq="ME", key='date'), columns='city')
```
