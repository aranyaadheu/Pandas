```python
# parse the date column as a date

df = pd.read_csv("aapl.csv", parse_dates=['Date'], index_col='Date')
df.head(2)
```

```python
# Average price of aapl's stock in June, 2017
# "Close" column, which typically holds 
# the closing stock price for each day.
df.loc['2017-06'].Close.mean()
```

```python
# first two row from the year 2017
df.loc['2017'].head(2)
```

### (2) Select Date Range

```python
# from present to next days

df = df.sort_index()
df.loc['2017-01-01':'2017-01-07']
```

```python
df.loc['2017-01']
```

Resampling in pandas is a time-series operation that involves changing the frequency of your time-series data â€” either upsampling (more granular) or downsampling (less granular).

```python
# ME as month-end
df['Close'].resample('ME').mean().head()
```
